```python
import yfinance as yf
import datetime

# Define the stock ticker (e.g., 'TCS.BO' for Tata Consultancy Services on BSE)
ticker = 'TCS.BO'

# Define the time period (last 10 days)
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=10)

# Download the stock data for the last 10 days
stock_data = yf.download(ticker, start=start_date, end=end_date)

# Display the fetched data
print(stock_data)

# saving the data to a CSV file
stock_data.to_csv(f'{ticker}_last_10_days.csv')
```

```python
import yfinance as yf
import datetime
import pandas as pd

# List of some BSE stock tickers  (5 stock indicators)
tickers = ['TCS.BO', 'RELIANCE.BO', 'INFY.BO', 'HDFCBANK.BO', 'ITC.BO']

# Define the time period (last 10 days)
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=10)

# Download the stock data for the last 10 days for all tickers
stock_data = yf.download(tickers, start=start_date, end=end_date)

# Print the fetched data
print(stock_data)

# Optionally, save the data to a CSV file
stock_data.to_csv('bse_stocks_last_10_days.csv')

# If you want to save individual stocks data
for ticker in tickers:
    individual_data = stock_data.xs(ticker, level=1, axis=1)
    individual_data.to_csv(f'{ticker}_last_10_days.csv')
```

```python
import yfinance as yf
import pandas as pd
import datetime
import os

# Define the time period (last 10 days)
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=10)

# Load the CSV file containing the stock list
stock_list_df = pd.read_csv('/content/Equity.csv')  # Replace with your actual file path

# Create a directory to store the fetched data
output_dir = "bse_stock_data"
os.makedirs(output_dir, exist_ok=True)

# Iterate over each stock in the list
for index, row in stock_list_df.iterrows():
    security_id = row['Security Id']  # Using Security Id for yfinance ticker symbol
    ticker = f"{security_id}.BO"  # Construct the yfinance ticker

    try:
        # Fetch the stock data for the last 10 days
        stock_data = yf.download(ticker, start=start_date, end=end_date)

        # Check if data is returned
        if not stock_data.empty:
            # Save the data to a CSV file named after the security id
            output_file = os.path.join(output_dir, f"{security_id}_last_10_days.csv")
            stock_data.to_csv(output_file)
            print(f"Data for {security_id} saved to {output_file}")
        else:
            print(f"No data found for {security_id}")

    except Exception as e:
        print(f"Failed to fetch data for {security_id}: {e}")

print("Data fetching completed.")
```
