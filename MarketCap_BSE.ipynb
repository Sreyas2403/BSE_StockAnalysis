```python
import yfinance as yf
import pandas as pd

# Load the CSV file containing the stock list
stock_list_df = pd.read_csv('/content/Equity.csv')  #keep your file path

# Define market cap thresholds (These are approximate and can vary by region)
large_cap_threshold = 10_000_000_000  # $10 billion
mid_cap_threshold = 2_000_000_000     # $2 billion

# List to store the market cap and type
market_cap_info = []

# Iterate over each stock in the list
for index, row in stock_list_df.iterrows():
    security_id = row['Security Id']  # Using Security Id for yfinance ticker symbol
    ticker = f"{security_id}.BO"  # Construct the yfinance ticker

    try:
        # Fetch stock data using yfinance
        stock = yf.Ticker(ticker)
        stock_info = stock.info

        # Get market cap and share price
        market_cap = stock_info.get('marketCap', None)
        last_price = stock_info.get('previousClose', None)

        # Classify the market cap
        if market_cap:
            if market_cap >= large_cap_threshold:
                cap_type = 'Large Cap'
            elif market_cap >= mid_cap_threshold:
                cap_type = 'Mid Cap'
            else:
                cap_type = 'Small Cap'
        else:
            cap_type = 'Unknown'

        # Append the information to the list
        market_cap_info.append({
            'Security Id': security_id,
            'Last Price': last_price,
            'Market Cap': market_cap,
            'Cap Type': cap_type
        })

    except Exception as e:
        print(f"Failed to fetch data for {security_id}: {e}")

# Convert the list to a DataFrame
market_cap_df = pd.DataFrame(market_cap_info)

# Display the result
print(market_cap_df)

#  save the DataFrame to a CSV file
market_cap_df.to_csv('bse_market_cap_info.csv', index=False)
```
